```python
with open('/Users/dgrubis/Desktop/DS5500/repo/DS5500_Player_Tracking_and_Identification_NBA/data/0021500440.json') as r:
  jsonData = json.load(r)
```

```python
rows = []
moment_id = 0
#each event, I *think* is a possession or similar event found in play by play
#TODO: look into how pbp splits events vs. SportVu here
for i in range(len(jsonData['events'])):
    event_id = int(jsonData['events'][i]['eventId'])
    #moment is every frame captured at 25 fps
    for moment in jsonData['events'][i]['moments']:
        moment_id += 1
        quarter = moment[0]
        mili_seconds = moment[1]
        game_clock = moment[2]
        shot_clock = moment[3]
        #entity = player or ball
        for entity in moment[5]:
            team_id = entity[0]
            player_id = entity[1]
            x = entity[2]
            y = entity[3]
            z = entity[4]
            row = [event_id, moment_id, quarter, mili_seconds, game_clock, shot_clock, team_id, player_id, x, y, z]
            rows.append(row)

rows = np.array(rows)
df = pd.DataFrame(rows, columns = ['event_id', 'moment_id', 'quarter', 'mili_seconds', 'game_clock', 
                                  'shot_clock', 'team_id', 'player_id', 'x', 'y', 'z'])
```
